보라색 노드로 표시된 그룹은 13명, 주황색 노드로 표시된 그룹은 14명입니다.
따라서, 최대 그룹의 인원은 14명입니다.

img5.png

나늘 그룹의 수가 4개 이므로, 그림과 같이 3개의 모든 간선을 끊어서 4개의 그룹으로 나눌 수 있습니다.
최대 그룹의 인원은 9명입니다.

```python
from collections import defaultdict

def solution(k, num, links):
    def Sum(i):
        return siteN[i] + sum(Sum(child) for child in graph[i])

    def Del(i):
        siteN[i] = 0
        for child in graph[i]:
            if siteN[child]: Del(child)

    siteN = defaultdict(int)
    graph = defaultdict(list)
    for i, n in zip(range(len(num)), num):
        siteN[i] = n
    for i, link in zip(range(len(num)), links):
        left, right = link
        if left > -1:
            graph[i].append(left)
        if right > -1:
            graph[i].append(right)
    Mean = sum(num) / k
    sumLst = [(i, Sum(i)) for i in range(len(num))]
    nop = []
    while len(nop) < k - 1:
        i, dif = sorted(map(lambda x: (x[0], abs(Mean - x[1])), sumLst), key=lambda x: x[1])[0]
        nop.append(Sum(i))
        Del(i)
        sumLst = [(i, Sum(i)) for i in range(len(num))]
    else:
        nop.append(sorted(sumLst, key=lambda x: x[1])[-1][1])
    return max(nop)
```
