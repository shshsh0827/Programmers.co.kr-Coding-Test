따라서 처음 주어진 표의 상태와 최종 표의 상태를 비교하면 다음과 같습니다.

```python
def solution(n, k, cmd):
    def U(pos, x):
        times = 0
        while times < x:
            pos -= 1
            if pos not in delete:
                times += 1
        return pos

    def D(pos, x):
        times = 0
        while times < x:
            pos += 1
            if pos not in delete:
                times += 1
        return pos

    def C(pos):
        delete.append(pos)
        while True:
            if pos == n - 1:
                while True:
                    pos -= 1
                    if pos not in delete:
                        return pos
            pos += 1
            if pos not in delete:
                return pos

    def Z():
        delete.pop()
    
    delete = []
    pos = k
    for c in cmd:
        if c == 'C':
            pos = C(pos)
        elif c == 'Z':
            Z()
        else:
            c, x = c.split()
            if c == 'U':
                pos = U(pos, int(x))
            else:
                pos = D(pos, int(x))
    result = ['O'] * n
    for i in delete:
        result[i] = 'X'
    return ''.join(result)
```
